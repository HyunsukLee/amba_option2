Compute the average and max price charged.  Store the results in variables called avg_price and max_price

```python
#Write your code here
import pandas as pd
df_data = pd.read_csv("/content/sample_data/Purchase_Data.csv")

totalPrice = sum(df_data.price)
cntPrice = len(df_data.price)

avg_price  = totalPrice / cntPrice
max_price = max(df_data.price)

print("mean : ", df_data['price'].mean() )
print("avg_price : ", avg_price)
print("max_price : ", max_price)
```
